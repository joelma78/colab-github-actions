```python
import pandas as pd

dados = {
    "Aluno": ["Ana", "Bruno", "Carlos"],
    "Nota": [8.5, 7.0, 9.0]
}

df = pd.DataFrame(dados)
df
```

```python
media = df["Nota"].mean()
print("Média das notas:", media)
```

```python
# Adicionando nova coluna baseada na nota
df["Resultado"] = df["Nota"].apply(
    lambda x: "Aprovado" if x >= 7 else "Reprovado"
)

# Exibindo o DataFrame final
df
```
